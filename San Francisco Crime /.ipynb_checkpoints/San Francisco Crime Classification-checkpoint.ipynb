# San Francisco Crime Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Explore a Data set

<p style="font-size=16px">This dataset contains incidents derived from SFPD Crime Incident Reporting system. The data ranges from 1/1/2003 to 5/13/2015.</p>
### Data fields
<ul>
<li><strong>Dates</strong> - timestamp of the crime incident</li>
<li><strong>Category</strong> - category of the crime incident.</li>
<li><strong>Descript</strong> - detailed description of the crime incident</li>
<li><strong>DayOfWeek</strong> - the day of the week</li>
<li><strong>PdDistrict</strong> - name of the Police Department District</li>
<li><strong>Resolution</strong> - how the crime incident was resolved</li>
<li><strong>Address</strong> - the approximate street address of the crime incident </li>
<li><strong>X</strong> - Longitude</li>
<li><strong>Y</strong> - Latitude</li>
</ul>

Data Source : https://data.sfgov.org/Public-Safety/SFPD-Incidents-from-1-January-2003/tmnf-yvry

```python
# Load .csv data
SanFranciscoCrimeData=pd.read_csv('../input/SanFranciscoCrimeData.csv',low_memory=False)
```

```python
# Take a look of the data
SanFranciscoCrimeData.head()
```

### Which kind of crime happened more often ?

```python
SanFranciscoCrimeData.Category.value_counts(normalize=True)
```

```python
# For better visualization purpose , import seaborn library
import seaborn as sns
fig=plt.figure(1,figsize=(12,8))
ax1 = fig.add_subplot(1,1,1)
SanFranciscoCrimeData.Category.value_counts(normalize=True).plot(kind='bar')
plt.xlabel('Category of Crime')
plt.ylabel('Frequency')
plt.title('Occurance of Crime by Category')
```

### Most of the Crime Committed on what time ?

```python
# Add column containing day of week expressed in integer
dayOfWeek = {
    'Monday':0,
    'Tuesday':1,
    'Wednesday':2,
    'Thursday':3,
    'Friday':4,
    'Saturday':5,
    'Sunday':6
}
SanFranciscoCrimeData['DOW'] = SanFranciscoCrimeData.DayOfWeek.map(dayOfWeek)

# Add column containing time in hour of day
SanFranciscoCrimeData['Hour'] = pd.to_datetime(SanFranciscoCrimeData.Dates).dt.hour
```

```python
SanFranciscoCrimeData.Hour.value_counts(normalize=True)
```

```python
plt.figure(1,figsize=(12,8))
SanFranciscoCrimeData.Hour.value_counts(normalize=True).plot(kind='bar')
plt.xlabel('Time of Day')
plt.ylabel('Crime')
plt.title('Occurance of Crime by Hours')
```

### Plot the Time vs Day heat graph for all Kind of Crimes

```python
plt.figure(figsize=(12,8))
plt.hist2d(
    SanFranciscoCrimeData.Hour.values,
    SanFranciscoCrimeData.DOW.values,
    bins=[24,7],
    range=[[-0.5,23.5],[-0.5,6.5]]
)
plt.xticks(np.arange(0,24,3))
plt.xlabel('Time of Day')
plt.yticks(np.arange(0,7),['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
plt.ylabel('Day of Week')
plt.gca().invert_yaxis()
plt.title('Occurance of Crime by Time and Day All')
```

### Plot the Time vs Day heat graph for LARCENY/THEFT category of crime

```python
plt.figure(figsize=(12,8))
plt.hist2d(
    SanFranciscoCrimeData[SanFranciscoCrimeData.Category=='LARCENY/THEFT'].Hour.values,
    SanFranciscoCrimeData[SanFranciscoCrimeData.Category=='LARCENY/THEFT'].DOW.values,
    bins=[24,7],
    range=[[-0.5,23.5],[-0.5,6.5]]
)
plt.xticks(np.arange(0,24,3))
plt.xlabel('Time of Day')
plt.yticks(np.arange(0,7),['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
plt.ylabel('Day of Week')
plt.gca().invert_yaxis()
plt.title('Occurance of LARCENY/THEFT by Time and Day All')
```
